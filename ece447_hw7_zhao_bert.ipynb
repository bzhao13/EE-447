# 1. Root Locus

For each of the following plants, plot the root locus of the system
obtained using no compensator. Determine the values of: imaginary axis
crossings, points at which complex-conjugate pairs join on the real
axis, and asymptotes (intercept and angle) for zeros and infinity.

- $K G(s) = \frac{K(s+1)}{(s+2)(s+3)}$
- $K G(s) = \frac{K(s^2+1)}{(s+3)(s^2+s+1)}$
- $K G(s) = \frac{K}{(s+1)(s+2)(s^2+1)}$


##System 1

System 1 does not have any imaginary axis crossings. The pole at $-3$ goes to $-\infty$ as $K \rightarrow \infty$. and the pole at $-2$ goes to the zero at $-1$ as $K \rightarrow \infty$.

There are no complex conjugate pairs in system one so there is no distinct point at which they join on the real axis. The poles start on the real axis and remain on the real axis as $K \rightarrow \infty$.

Solving for the real axis asymptote, we get:

$$
\sigma_0 = \frac{\sum \mathrm{finite\;poles} - \sum \mathrm{finite\;zeros}}{\# \mathrm{finite\;poles} - \# \mathrm{finite\;zeros}} = \frac{[-2-3]-[-1]}{2-1} = -4
$$

Solving for the angles of the asymptoes, we get:

$$
k = \# \mathrm{finite\;poles} - \# \mathrm{finite\;zeros} - 1 = 2 - 1 - 1 = 0
$$

$$
\theta = \frac{(2k+1)\pi}{\# \mathrm{finite\;poles} - \# \mathrm{finite\;zeros}} = \frac{\pi}{2-1} = \pi
$$

##System 2

System 2 does not have any imaginary axis crossings at a defined value of $K$. However, the poles at $-\frac{1}{2} \pm \frac{\sqrt2}{2}i$ will approach the imaginary axis at $\pm i$ as $K \rightarrow \infty$. The pole at $-3$ goes to $-\infty$ as $K \rightarrow \infty$.

The complex conjugate pair of $-\frac{1}{2} \pm \frac{\sqrt2}{2}i$ move away from the real axis as $K \rightarrow \infty$ so there is no distinct point at which they join on the real axis. The pole at $-3$ stays on the real axis as $K \rightarrow \infty$.

Solving for the real axis asymptote, we get:

$$
\sigma_0 = \frac{\sum \mathrm{finite\;poles} - \sum \mathrm{finite\;zeros}}{\# \mathrm{finite\;poles} - \# \mathrm{finite\;zeros}} = \frac{[-3+(-\frac{1}{2} + \frac{\sqrt2}{2}i)+(-\frac{1}{2} - \frac{\sqrt2}{2}i)]-[0]}{3-2} = -4
$$

Solving for the angles of the asymptoes, we get:

$$
k = \# \mathrm{finite\;poles} - \# \mathrm{finite\;zeros} - 1 = 3 - 2 - 1 = 0
$$

$$
\theta = \frac{(2k+1)\pi}{\# \mathrm{finite\;poles} - \# \mathrm{finite\;zeros}} = \frac{\pi}{3-2} = \pi
$$

##System 3

System 3 only has imaginary axis crossings at $\pm i$ when $K=0$. The poles at $-1$ and $-2$ meet at about $-1.6$ before they go to $-\infty$ as $K \rightarrow \infty$.

The all poles move away from the real axis as $K \rightarrow \infty$ so there is no distinct point at which they join on the real axis. The poles at $-1$ and $-2$ start on the real axis and meet at about $-1.6$ before they go to $-\infty$ as $K \rightarrow \infty$.

Solving for the real axis asymptote, we get:

$$
\sigma_0 = \frac{\sum \mathrm{finite\;poles} - \sum \mathrm{finite\;zeros}}{\# \mathrm{finite\;poles} - \# \mathrm{finite\;zeros}} = \frac{[-1+-2+-i+i]-[0]}{4-0} = \frac{-3}{4}
$$

Solving for the angles of the asymptoes, we get:

$$
k = \# \mathrm{finite\;poles} - \# \mathrm{finite\;zeros} - 1 = 4 - 0 - 1 = 3
$$

$$
\theta = \frac{(2k+1)\pi}{\# \mathrm{finite\;poles} - \# \mathrm{finite\;zeros}} = \frac{\pi}{4-0} = \frac{\pi}{4}, \frac{3\pi}{4}, \frac{5\pi}{4}, \frac{7\pi}{4}
$$

```python
def find_K(r, real, imag, perTol):
  i = 0
  while i < r.size:
    realNum = abs(np.real(r[i]))
    imagNum = np.imag(r[i])
    if realNum > (1-perTol) * abs(real) and realNum < (1+perTol) * abs(real):
      if imagNum < -(1-perTol) * imag and imagNum > -(1+perTol) * imag:
        return i
      elif imagNum > (1-perTol) * imag and imagNum < (1+perTol) * imag:
        return i
    i += 1
  
  return 1
```

# 3. Control of a Ducted Fan Wing

Check out [this video](https://www.youtube.com/watch?v=ZFb5kFpgCm4) of a ducted fan. 

The transfer function for the pitch subsystem is

$$
G(s) = \frac{\theta(s)}{\tau(s)} = \frac{r}{J s^2 + bs + mgl}
$$

where $\theta(s)$ is the pitch and $\tau(s)$ is the input torque. The
values of the constants were measured and found to be $g = 9.81
\;m/s^2$, $m=1.5 \; kg$, $b = 0.05 kg/s$, $l = 0.05 m$, $J = 0.0475 kg
m^2$ and $r = 0.25 m$.

Recall that a PID controller has the form

$$
G_C(s) = \frac{K(s-z_1)(s-z_2)}{s}
$$

Find positions for the zeros and a value for $K$ so that the step response has a settling time of $0.1$ seconds and an overshoot of 10%. Plot the root locus for your system and show indicate where value of $K$ you used is.

Find the sensitivity of the percent overshoot in the closed loop system to changes in $b$. Make a plot of percent overshoot vs $b$ where $b$ goes from 50% of the value given above to 150% of the value given above.



We can plug in the given values to obtain the following $G(s)$:

$$G(s) = \frac{0.25}{0.0475s^2+0.05s+0.73575}$$

To find the controller we want to use, we can find the $\zeta$ and $\omega$ that satisfy the percent overshoot and settling time.

$$
\frac{4}{\zeta\omega} = 0.1 \Rightarrow \;\; \zeta\omega=40
$$

$$
\frac{\zeta\pi}{\beta} = - \ln (0.1)
$$

We get:

$$\zeta = 0.5912$$

$$\omega = 67.6641$$

This gives a characteristic equation of:

$$s^2+80s+4578.44$$

Which we use to place our zeros. The zeros are:

$$z = -40 \pm 54.58i$$


The zeros will need to be from the controller, so we can design an approximatly equivalent controller using:

$$G_c(s) = \frac{s^2+78s+4330}{s}$$

This gives us a transfer function that we can use to plot with:

$$T(s) = \frac{0.25(s^2+78s+4330)}{0.0475s^3+0.05s^2+0.73575s}$$

From the plots below, we can see that we can move the poles to approximatly $-40.05 \pm 57.30i$ with a gain of about $K=131.79$. We can then look at the time domain plots to verify that the percent overshoot and settling time are met.

After finding a value of K, we get the following complete transfer function:

$$T(s) = \frac{KG_cG}{1+KG_cG} = \frac{K0.25(s^2+78s+4330)}{Js^3+(b+rK)s^2+(mgl+rKz_1+rKz_2)s + rKz_1z_2}$$

Which we can substitute our values to get:

$$T(s) = \frac{131.79 \cdot 0.25 \cdot (s^2+78s+4330)}{(s+614.58)(s^2 + 80.1s + 4887.29)}$$

To get our third order system to match the approximation using a second order approximation, we can factor the denominator as follows:

$$T(s) = \frac{n(s)}{(s+a)(s^2+2\zeta\omega s + \omega^2)} = 
\frac{n(s)}{s^3+(2\zeta\omega +a)s^2+(\omega^2+2\zeta\omega a)s + a\omega^2}$$

And then we can do coeffecient matching to find $\zeta$ and $\omega$.

$$\frac{b+rK}{J} = 2\zeta\omega +a$$

$$ \zeta = (\frac{b+rK}{J}-a)\frac{1}{2\omega}$$

$$ a\omega^2 = \frac{rKz_1z_2}{J} $$

$$ \omega = \sqrt{\frac{rKz_1z_2}{Ja}}$$

Substituting with the values used in the controller, we get:

$$ \omega = 74.267 $$

$$ \zeta = 0.142b + 0.532 $$

We can then find percent overshoot:

$$\text{P.O.} = e^{\frac{-\zeta \pi}{\beta}} = e^{\frac{-(0.142b + 0.532) \pi}{\sqrt{1-(0.142b + 0.532)^2}}}$$

And the sensitivity:

$$S^{PO}_b = \frac{\partial \text{P.O.}}{b} \cdot \frac{b}{\text{P.O.}} = -\frac{0.142 \pi b}{(-0.02b^2 - 0.15b + 0.72)^{\frac{3}{2}}}$$

The plot for $S^{PO}_b$ varying $b$ from 50% to 150% is shown below. We can see that the percent overshoot is not very sensitive to $b$. This is because the relativily high gain makes the controller do most of the work of the system. Varying the $b$ value from 50% to 150% in the plant does not do much to affect the overshoot unless a new controller with a much smaller gain is used.
