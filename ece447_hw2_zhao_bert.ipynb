and should use the approximation to the definite integral as a sum of the areas of rectangles with base ranging from `x` to `x+delta` and height `f(x)` as x goes from `x1` to `x2` in steps of size `delta`. Show the results of testing your function to compute approximations of the integral of `np.sin` from `0` to $\pi$ with `delta = 1, 0.1, 0.01, 0.001`. ----

````python
def approx_integral(f, x1, x2, delta):
  sum = 0
  start = x1
  end = x2

  while start < end:
    sum += (delta * f(start + delta))
    start += delta
  
  return sum

def f(x):
  return np.sin(x)

delta = [1, 0.1, 0.01, 0.001]

for x in delta:
  print("The approximate integral of sin(x) from x = 0 to x = \u03C0 is:", approx_integral(f, 0, np.pi, x), "for delta =", x)
````

**a)** Find an open loop control law (a constant) for $u$ that makes the steady state  $x_{ss}$ equal to a desired state $r$.

**b)** Find $x_{ss}$ in terms of $\hat k$ and $k$. Explain what happens to $x_{ss}$ upon a 10% change in $k$. Plot $x_{ss}$ as a function of $k$ it goes from $k=0.9 \hat k$ to $k= 1.1 \hat k$.

**c)** Find a proportional controller for $u$ that makes the steady state $x_{ss}$ equal to a desired state $x_{des}$.

**d)** Find the steady state velocity $x_{ss}$ in terms of $\hat k$, $k$, and $K_P$. Explain what happens to $x_{ss}$ when $K_P$ increases and what happens as it goes to infinity. 

**e)** Suppose $r = 1$ and $\hat k = 10$ and $k = 9$. Find a value for $K_P$, call it $K_P^*$, that makes $|x_{ss} - x_{des}| < 0.01$. Make a plot of $x_{ss}$ as $K_P$ varies from $0$ to $K_P^*$.  

##a) For steady state, 

$$\dot x = 0 = -x+ku$$
$$0 = -x_{ss}+ku$$
$$u=\frac{x_{ss}}{k}$$

Where $x_{ss}$ is our desired state, $r$

##b) For steady state solution, we know that:

$$\dot x = 0 = -x_{ss} + ku$$

We first replace $x_{ss}$ with $r$ for our desired state and $k$ with $\hat k$ to represent our approximation for $k$.

$$u = \frac{x_{ss}}{k} = \frac{r}{\hat k}$$

We can then substitute $u$ to get:

$$ 0 = -x_{ss} + k \cdot \frac{r}{\hat k}$$

And we can rearrange the equation to get $x_{ss}$ as a function of the ratio of $k$ and $\hat k$.

$$x_{ss} = \frac{k}{\hat k} \cdot r$$

This means that there is a proportional relationship between the true steady state, $x_{ss}$, and the desired state, $r$, based on the accuracy of $k$.

````python
k = np.linspace(0.9, 1.1, 100)

def x_ss (k, k_hat, r):
  return k / k_hat * r

plt.plot(k, x_ss(k, 10, 1))
plt.xlabel('$\hat k$')
plt.ylabel('$x_{ss}$')
plt.title('$x_{ss}$ as a function of $\hat k$ with $r=1$')
````

##c) We know that a generic proportional controller follows the model:

$$u=p_{o}+K_{P}e$$

Where $e$ is the error represented by $r-x_{ss}$ and $K_{p}$ is the proportional gain.

If $x_{ss}$ equals our desired state $r$, then the error, $e$ equals $0$.

$$u=p_{o}$$

Substituting back into the steady state equation, we get:

$$0=-r+\hat kp_{o}$$

And solving for $p_{o}$, we get:

$$p_{o} = \frac{r}{\hat k}$$

Where $p_{o}$ can be the proportional control input, $u$, to the system.

##d) When there is non-zero error, our system needs to account for $K_{p}$ which can be modeled by substituting in the proportional feedback.

$$\dot x = 0 = -x_{ss} + k [p_{o} + K_{P}(r-x_{ss})]$$

$$0 = -x_{ss} + k [\frac{r}{\hat k} + K_{P}(r-x_{ss})]$$

The equation can then be rearranged for $x_{ss}$

$$x_{ss} =  k [\frac{r}{\hat k} + K_{P}(r-x_{ss})]$$
$$x_{ss} =  k \cdot \frac{r}{\hat k} + k \cdot K_{P}(r-x_{ss})$$
$$x_{ss} =  k \cdot \frac{r}{\hat k} + k \cdot K_{P} \cdot r - k \cdot K_{P} \cdot x_{ss}$$
$$x_{ss} + k \cdot K_{P} \cdot x_{ss}=  k \cdot \frac{r}{\hat k} + k \cdot K_{P} \cdot r $$
$$x_{ss} (1 + k \cdot K_{P}) =  r (\frac{k}{\hat k} + k \cdot K_{P})$$
$$\hat k \cdot x_{ss} (1 + k \cdot K_{P}) = \hat k \cdot r (\frac{k}{\hat k} + k \cdot K_{P})$$
$$x_{ss} (\hat k + \hat k \cdot k \cdot K_{P}) = r (k + \hat k \cdot k \cdot K_{P})$$

$$x_{ss} = \frac{(k + \hat k \cdot k \cdot K_{P})}{(\hat k + \hat k \cdot k \cdot K_{P})} \cdot r$$

We can then replace $\hat k \cdot k$ with constant c to get:

$$x_{ss} = \frac{(k + c \cdot K_{P})}{(\hat k + c \cdot K_{P})} \cdot r$$

And we can see that as $K_{p} \rightarrow \infty, x_{ss} \rightarrow r$.  

##e) For the given quantities, we know that:

$$|x_{ss} - 1| < 0.01$$
$$0.99 < x_{ss} < 1.01$$

The expression for the system is also given as:

$$x_{ss} = \frac{(9 + 90 \cdot K_{P})}{(10 + 90 \cdot K_{P})} \cdot 1$$

If we take $x_{ss} = 1.01$, then we can calculate $K_{P}^{*} = -1.22$, and if we take $x_{ss} = 0.99$, then we can calculate $K_{P}^{*} = 1$

````python
def system (k, k_hat, r, Kp):
  return (k + (k * k_hat * Kp)) / (k_hat + (k * k_hat * Kp))

Kp_1 = np.linspace(0, 1, 1000)

plt.figure(figsize=(15,15))
plt.figure(1)
plt.plot(Kp_1, system(9, 10, 1, Kp_1))
plt.title('$x_{ss}$ as a function of $K_{P}$', fontdict= {'fontsize': 20})
plt.xlabel('$K_{P}$', fontdict= {'fontsize': 20})
plt.ylabel('$x_{ss}$', fontdict= {'fontsize': 20})
````

$$ A = \left(
\begin{array}{cc}
 1 & -1 \\
 -2 & 1 \\
\end{array}
\right) $$

There exists distinct eignevalues and eigenvectors for the matrix $A=\begin{bmatrix}1 & -1\\-2 & 1\end{bmatrix}$ that satisfy the equation $Av=\lambda v$.

The equation can be rearranged as such:

$$Av- \lambda v = 0$$
$$(Av-\lambda Iv) = 0$$
$$(A-\lambda I)v = 0$$

For solutions of non-zero values of $v$, there must be solutions for $|(A-\lambda I)| = 0$.

Then:

$$|(A-\lambda I)| = |\begin{bmatrix}1 & -1 \\ -2 & 1\end{bmatrix} - \begin{bmatrix}\lambda & 0 \\ 0 & \lambda\end{bmatrix}| = 0$$

$$|\begin{bmatrix}1-\lambda & -1 \\ -2 & 1-\lambda\end{bmatrix}| = 0$$

The determinate of the matrix is:

$$(1-\lambda)^2 - 2 = 0$$

And that gives the characteristic equation:

$$\lambda^{2}-2\lambda-1=0$$

Which solve for eigenvalues:

$$\lambda = \frac{2 \pm \sqrt{2^2-4\cdot1\cdot-1}}{2} $$
$$\lambda_1=1+\sqrt{2}, \lambda_2=1-\sqrt{2}$$

Solving for $v_1$:

$$\begin{bmatrix}1-\lambda_1 & -1 \\ -2 & 1-\lambda_1\end{bmatrix} \cdot v_1 = 0$$
$$\begin{bmatrix}-\sqrt{2} & -1 \\ -2 & -\sqrt{2}\end{bmatrix} \cdot \begin{bmatrix} v_{1,1} \\ v_{1,2}\end{bmatrix} = 0$$

$$-\sqrt{2}v_{1,1}+-v_{1,2} = 0$$
$$-2v_{1,1}+-\sqrt{2}v_{1,2} = 0$$
$$-\sqrt{2}v_{1,1} = v_{1,2}$$
$$v_1 = \begin{bmatrix}-\frac{1}{\sqrt{2}} \\ 1 \end{bmatrix}$$

Solving for $v_2$:

$$\begin{bmatrix}1-\lambda_2 & -1 \\ -2 & 1-\lambda_2\end{bmatrix} \cdot v_2 = 0$$
$$\begin{bmatrix}\sqrt{2} & -1 \\ -2 & \sqrt{2}\end{bmatrix} \cdot \begin{bmatrix} v_{2,1} \\ v_{2,2}\end{bmatrix} = 0$$

$$\sqrt{2}v_{2,1}+-v_{2,2} = 0$$
$$-2v_{1,1}+\sqrt{2}v_{1,2} = 0$$
$$\sqrt{2}v_{2,1} = v_{2,2}$$
$$v_2 = \begin{bmatrix}\frac{1}{\sqrt{2}} \\ 1 \end{bmatrix}$$

Final eigenvalues and eigenvectors are:

$$\lambda_1=1+\sqrt{2}, \lambda_2=1-\sqrt{2}$$
