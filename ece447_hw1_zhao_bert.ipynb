##System 1: Electric Stovetop
- The input is a temperature knob and the output is heat energy.
- The feedback system makes sure that the rate of heat energy output does not excede the amount nessesary to reach the desired input temperature.
- The controller is impletmented with a physical temerature sensor to measure the temperature of the hot plate.
- Disturbances are most likely putting a pan on the hot plate and food of lower temperatures absorbing heat energy which would change the temperature of the hot plate.
- There are two burners next to each other, and when both are on, heat from the one nearby can introduce noise if each temperature sensor is not insulated from the nearby burners.
- The reference input is a knob which is just used to get a reference temperature from the user.
- The most important performance requirement is speed, since it is most convienent when cooking to get to the ideal temperature in a relativily quick ammount of time. Accuracy is not as important because slight variations of a few degrees do not impact cooking as much as overall cooking time.


##System 2: Toilet
- The input is current water level in the tank, and the output is the rate of water flow into the tank through the fill valve. The input can also be thought of as the user when they flush, which lowers the current water level in the tank.
- The feedback system ensures that the water level in the tank does not excede a specific level.
- The controller is a mechanical floater. When the water level rises to the desired point, the floater will close off the fill valve.
- Most common disturbance is general wear and tear to the refil valve. A damaged valve can cause the water to not fill to the desired level. Other less common disturbances could be a strong storm which can also alter the water level in the tank.
- Once source of noise could be vibrations to the floater, this could cause very slight variations which can cause the fill valve to close at varying water levels.
- The reference input which is the desired water level, is determined by the manufacturer and it is so that the water does not overflow out of the tank.
- Accuracy is the most important aspect because it is important that the water does not overflow. Cost is also important since toilets are mass produced. Speed is less important since the most common case is only flushing a toilet once.

## 3) Basic python (10pts)
Create a python function to compute the normalzed sinc function $$\DeclareMathOperator{\sinc}{sinc}\DeclareMathOperator{\sincn}{sincn} \sincn(x) := \sinc(\pi x).$$  Then use your function to plot $\sincn(x)$ from -5 to 5.

```
import numpy as np
import matplotlib.pyplot as plt

def sincn(x):
  return np.sinc(np.pi * x)

t = np.linspace(-5, 5, 1000)
plt.plot(t, sincn(t))
plt.title("sincn(x)")
```
